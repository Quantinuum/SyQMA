# Basic usage of the simulator

Import necessary functions

```python
import numpy as np
from syqma import Simulator
```

Initialise the simulator. It is useful to specify an upper bound of the number of gates that will be used in the circuit. This is because the simulator uses a fixed-size array to store the gates. If this is not specified, the simulator will make a guess, which might be too small.

```python
n = 10

sim = Simulator(n)
```

Implement the circuit by adding gates as methods of the `Simulator` object. In this case, we are implementing a GHZ circuit on n qubits.

```python
sim.h(0)
for i in range(n - 1):
    sim.cnot(i, i + 1)
```

We can check the stabiliser tableau that is being tracked.

```python
print(sim.pauli_span())
```

We can add measurements on some specified qubits, in this case on all the qubits. This will return the result of the measured `Z` observable in the form of two lists of phases and errors that arise from combining stabiliers to obtain the desired observable. If no such solution exists, the phase will be 1 and the error terms will be all 0.

```python
measured_qubits = np.array(list(range(n)))
output = sim.measure_all(measured_qubits)
```

We can now display the whole analysed circuit.

```python
print(sim.draw_circuit())
```

Since we only ran a noiseless simulation, we can check that the phase is 1 and the error terms are all 0.

```python
print("Solution phases:")
print(output.span.phases)
print()
print("Solution error terms:")
print(output.span.list_error_channels)
```

Now for the more interesting case, we can add noise to the simulation. We will still use the GHZ circuit, but we will add Pauli noise after every gate. Bear in mind that we do not need to specify the actual magnitudes of the Pauli terms in the error channel, since the simulator treats them symbolically.

```python
sim = Simulator(n)

sim.h(0)
sim.pauli_error_1(0)
for i in range(n - 1):
    sim.cnot(i, i + 1)
    sim.pauli_error_2((i, i + 1))
    
measured_qubits = np.array(list(range(n)))
output = sim.measure_all(measured_qubits)
```

Let us print the circuit again, which now includes the noise.

```python
print(sim.draw_circuit())
```

Let us now define an error channel in order to evaluate our expectation value. We initialise errors unformly based on a normal distribution.

```python
p0 = 1e-3
std_dev_scale = 0.0

p_base_1q = np.random.normal(p0, std_dev_scale * p0, 3)
errors_1q = np.hstack((1 - np.sum(p_base_1q), p_base_1q))
p_base_2q = np.random.normal(p0, std_dev_scale * p0, 15)
errors_2q = np.hstack((1 - np.sum(p_base_2q), p_base_2q))
```

Finally, to evaluate our desired expactation value with this noise channel, we need to define all the applied error channels, and then call the expectation value method.

```python
exp_val = sim.evaluate_expectation_value(
    output,
    errors_1q,
    errors_2q,
)

print(f"Expectation value: {exp_val}")
```
