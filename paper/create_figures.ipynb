```python
import json
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
import rsmf
from scipy import stats

# default is 11
fontsize_legend = 10
fontsize = 14
```

Figure 1: Iceberg code Discard rate for FT |0> state preparation and syndrome measurements

```python
folder = "fig_1_iceberg_discard"
with open(f"{folder}/data.json", "r") as f:
    d = json.load(f)

n_datas = d["n_datas"]
discard_rates_original_prep = d["original_prep"]
discard_rates_half = d["half_prep"]
discard_rates_quarter = d["quarter_prep"]
discard_rates_original_syndrome = d["original_syndrome"]
discard_rates_prep_syndrome = d["prep_syndrome"]

colors = plt.get_cmap("tab10")
markers = ["o", "^", "s", "D", "X", "v"]
markersize = 0
linewidth = 2.5

formatter = rsmf.setup(r"\documentclass[a4paper,12pt,unpublished,noarxiv]{quantumarticle}")
fig = formatter.figure(width_ratio=1.8)

plt.plot(n_datas, discard_rates_prep_syndrome, linewidth=linewidth, linestyle="-.", marker=markers[3], markersize=markersize, markeredgecolor="black", color=colors(4), label="One-branch prep. + Meas.")
plt.plot(n_datas, discard_rates_original_syndrome, linewidth=linewidth, linestyle="--", marker=markers[2], markersize=markersize, markeredgecolor="black", color=colors(8), label="Syndrome measurement")
plt.plot(n_datas, discard_rates_quarter, linewidth=linewidth, linestyle=":", marker=markers[1], markersize=markersize, markeredgecolor="black", color=colors(6), label="Four-branch preparation")
plt.plot(n_datas, discard_rates_original_prep, linewidth=linewidth, linestyle="-", marker=markers[0], markersize=markersize, markeredgecolor="black", color=colors(9), label="One-branch preparation")

plt.xlabel("Number of physical qubits ($k + 2$)", fontsize=fontsize)
plt.ylabel("Discard rate", fontsize=fontsize)
plt.xscale("log")
plt.xticks(n_datas, [str(n) for n in n_datas], fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.legend(fontsize=fontsize_legend + 1)
plt.grid()
plt.xlim(n_datas[0], n_datas[-1])
plt.tight_layout()
plt.savefig(f"{folder}/{"_".join(folder.split("_")[2:])}.pdf")
plt.savefig(f"figures/{"_".join(folder.split("_")[2:])}.pdf")
plt.show()
```

Figure 2: Steane code Z LER for FT |0> state preparation

```python
folder = "fig_2_steane_ler"
with open(f"{folder}/data.json", "r") as f:
    d = json.load(f)

p0 = d["p0"]
free = d["free"]
postselected = d["postselected"]
corrected = d["corrected"]
discard = d["discard"]

# number of points for fitting
n_points_fit = len(p0) // 2

fit_z_free = stats.linregress(np.log(p0[:n_points_fit]), np.log(free[:n_points_fit]))
fit_z_postselected = stats.linregress(np.log(p0[:n_points_fit]), np.log(postselected[:n_points_fit]))
fit_z_corrected = stats.linregress(np.log(p0[:n_points_fit]), np.log(corrected[:n_points_fit]))
fit_discard = stats.linregress(np.log(p0[:n_points_fit]), np.log(discard[:n_points_fit]))

colors = plt.get_cmap("tab10")
linewidth = 2.5

formatter = rsmf.setup(r"\documentclass[a4paper,12pt,unpublished,noarxiv]{quantumarticle}")
fig = formatter.figure(width_ratio=1.8)

plt.plot(p0, discard, linewidth=linewidth, color=colors(0), label=rf"Discard rate $\to p^{{{fit_discard.slope:.0f} \, \pm \, {(fit_discard.stderr / 10 ** (int(np.log10(fit_discard.stderr)) - 1)):.0f} \times 10^{{{np.log10(fit_discard.stderr):.0f}}}}}$")
plt.plot(p0, free, linewidth=linewidth, color=colors(3), label=rf"Uncorrected LER $\to p^{{{fit_z_free.slope:.0f} \, \pm \, {(fit_z_free.stderr / 10 ** (int(np.log10(fit_z_free.stderr)) - 1)):.0f} \times 10^{{{np.log10(fit_z_free.stderr):.0f}}}}}$")
plt.plot(p0, corrected, linewidth=linewidth, color=colors(1), label=rf"Corrected LER $\to p^{{{fit_z_corrected.slope:.0f} \, \pm \, {(fit_z_corrected.stderr / 10 ** (int(np.log10(fit_z_corrected.stderr)) - 1)):.0f} \times 10^{{{np.log10(fit_z_corrected.stderr):.0f}}}}}$")
plt.plot(p0, postselected, linewidth=linewidth, color=colors(2), label=rf"Postselected LER $\to p^{{{fit_z_postselected.slope:.0f} \, \pm \, {(fit_z_postselected.stderr / 10 ** (int(np.log10(fit_z_postselected.stderr)) - 1)):.0f} \times 10^{{{np.log10(fit_z_postselected.stderr):.0f}}}}}$")

plt.xlabel("Physical error rate ($p$)", fontsize=fontsize)
plt.ylabel("Discard / Logical error rate", fontsize=fontsize)
plt.xscale("log")
plt.yscale("log")
plt.legend(fontsize=fontsize_legend + 1)
plt.grid()
plt.tight_layout()
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.xlim(left=1e-5, right=0.5)
ax = plt.gca()
ax.xaxis.set_minor_locator(ticker.LogLocator(subs='auto'))
ax.tick_params(axis='x', which='minor', length=3)
plt.savefig(f"{folder}/{"_".join(folder.split("_")[2:])}.pdf")
plt.savefig(f"figures/{"_".join(folder.split("_")[2:])}.pdf")
plt.show()
```

Figure 3: Steane H magic LER

```python
folder = "fig_3_steane_h_magic_ler"
with open(f"{folder}/data.json", "r") as f:
    exp_vals = json.load(f)

base_angle = np.pi / 4

all_p0 = np.array(exp_vals["all_p0"])
all_exp_val_z_free = np.array(exp_vals["exp_val_z_free"])
all_exp_val_z_postselected = np.array(exp_vals["exp_val_z_postselected"])
all_exp_val_z_corrected = np.array(exp_vals["exp_val_z_corrected"])
all_exp_val_x_free = np.array(exp_vals["exp_val_x_free"])
all_exp_val_x_postselected = np.array(exp_vals["exp_val_x_postselected"])
all_exp_val_x_corrected = np.array(exp_vals["exp_val_x_corrected"])
all_acceptance_probabilities = np.array(exp_vals["acceptance_probabilities"])

noiseless_exp_val = np.cos(base_angle)

all_ler_free = (1 - (all_exp_val_z_free + all_exp_val_x_free) * noiseless_exp_val) / 2
all_ler_postselected = (1 - (all_exp_val_z_postselected + all_exp_val_x_postselected) * noiseless_exp_val) / 2
all_ler_corrected = (1 - (all_exp_val_z_corrected + all_exp_val_x_corrected) * noiseless_exp_val) / 2

all_discard_probabilities = 1 - np.array(all_acceptance_probabilities)

n_fitted = 3

fit_free = stats.linregress(np.log(all_p0)[:n_fitted], np.log(all_ler_free)[:n_fitted])
fit_postselected = stats.linregress(np.log(all_p0)[:n_fitted], np.log(all_ler_postselected)[:n_fitted])
fit_corrected = stats.linregress(np.log(all_p0)[:n_fitted], np.log(all_ler_corrected)[:n_fitted])

fit_discard = stats.linregress(np.log(all_p0)[:n_fitted], np.log(all_discard_probabilities)[:n_fitted])

colors = plt.get_cmap("tab10")
markers = ["o", "^", "s", "D", "X", "v"]
markersize = 0
linewidth = 2.5

formatter = rsmf.setup(r"\documentclass[a4paper,12pt,unpublished,noarxiv]{quantumarticle}")
fig = formatter.figure(width_ratio=1.8)

plt.plot(all_p0, all_discard_probabilities, linewidth=linewidth, marker=markers[3], markersize=markersize, markeredgecolor="black", color=colors(0), label=rf"Discard rate $\to p^{{{fit_discard.slope:.0f} \, \pm \, {(fit_discard.stderr / 10 ** (int(np.log10(fit_discard.stderr)) - 1)):.0f} \times 10^{{{np.log10(fit_discard.stderr):.0f}}}}}$")
plt.plot(all_p0, all_ler_free, linewidth=linewidth, marker=markers[0], markersize=markersize, markeredgecolor="black", color=colors(3), label=rf"Uncorrected LER $\to p^{{{fit_free.slope:.0f} \, \pm \, {(fit_free.stderr / 10 ** (int(np.log10(fit_free.stderr)) - 1)):.0f} \times 10^{{{np.log10(fit_free.stderr):.0f}}}}}$")
plt.plot(all_p0, all_ler_corrected, linewidth=linewidth, marker=markers[2], markersize=markersize, markeredgecolor="black", color=colors(1), label=rf"Corrected LER $\to p^{{{fit_corrected.slope:.0f} \, \pm \, {(fit_corrected.stderr / 10 ** (int(np.log10(fit_corrected.stderr)) - 1)):.0f} \times 10^{{{np.log10(fit_corrected.stderr):.0f}}}}}$")
plt.plot(all_p0, all_ler_postselected, linewidth=linewidth, marker=markers[1], markersize=markersize, markeredgecolor="black", color=colors(2), label=rf"Postselected LER $\to p^{{{fit_postselected.slope:.0f} \, \pm \, {(fit_postselected.stderr / 10 ** (int(np.log10(fit_postselected.stderr)) - 1)):.0f} \times 10^{{{np.log10(fit_postselected.stderr):.0f}}}}}$")

plt.xlabel("Physical error rate ($p$)", fontsize=fontsize)
plt.ylabel("Discard / Logical error rate", fontsize=fontsize)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.xscale("log")
plt.yscale("log")
plt.legend(fontsize=fontsize_legend + 1)
plt.grid()
plt.tight_layout()
plt.xlim(left=1e-5, right=0.5)
ax = plt.gca()
ax.xaxis.set_minor_locator(ticker.LogLocator(subs='auto'))
ax.tick_params(axis='x', which='minor', length=3)
plt.savefig(f"{folder}/{"_".join(folder.split("_")[2:])}.pdf")
plt.savefig(f"figures/{"_".join(folder.split("_")[2:])}.pdf")
plt.show()
```

Figure 4: Steane syndrome analysis (smooth)

```python
folder = "fig_4_steane_syndromes"
with open(f"{folder}/data.json", "r") as f:
    d = json.load(f)

all_p0 = d["all_p0"]
bitstrings = d["bitstrings"]
probs = d["probs"]
evs = np.array(d["evs"])

colors = plt.get_cmap("viridis")
n_curves = len(bitstrings)
colors = [colors(j / max(n_curves - 1, 1)) for j in range(n_curves)]
markers = ["o", "^", "s", "D", "X", "v"]
markersize = 10
linewidth = 2

formatter = rsmf.setup(r"\documentclass[a4paper,12pt,unpublished,noarxiv]{quantumarticle}")
fig = formatter.figure(width_ratio=1.8, aspect_ratio=0.9)

uncertainty_threshold = 1e-1

ev_free = 0.0
ev_corrected = 0.0
ev_uncertain = 0.0

prob_free = 0.0
prob_corrected = 0.0
prob_uncertain = 0.0

noise_index = 0

for i in range(len(bitstrings) - 1, -1, -1):
    color = colors[i]
    alpha = min(1, 0.0 + min(1, -1 / np.log10(probs[i][0]) * 2))
    if evs[i, 0] == np.max(evs[:, 0]) or evs[i, 0] == np.min(evs[:, 0]):
        label = f"{bitstrings[i]}"
    else:
        label = None
    plt.plot(all_p0, evs[i], linewidth=linewidth, color=color, alpha=alpha, label=label)

    ev_free += probs[i][noise_index] * evs[i, noise_index]
    prob_free += probs[i][noise_index]
    ev_corrected += probs[i][noise_index] * abs(evs[i, noise_index])
    prob_corrected += probs[i][noise_index]
    if abs(evs[i, noise_index]) > uncertainty_threshold:
        ev_uncertain += probs[i][noise_index] * abs(evs[i, noise_index])
        prob_uncertain += probs[i][noise_index]

ev_uncertain /= prob_uncertain

ev_postselected = evs[-1, noise_index]
prob_postselected = probs[-1][noise_index]

print(f"Noise: {all_p0[noise_index]}, {noise_index = } / {len(all_p0)}\n")
print(f"Free:         prob = {prob_free:.16f}, ev = {ev_free:.16f}, ler = {(1 - ev_free):.16f}")
print(f"Corrected:    prob = {prob_corrected:.16f}, ev = {ev_corrected:.16f}, ler = {(1 - ev_corrected):.16f}")
print(f"Uncertain:    prob = {prob_uncertain:.16f}, ev = {ev_uncertain:.16f}, ler = {(1 - ev_uncertain):.16f}")
print(f"Postselected: prob = {prob_postselected:.16f}, ev = {ev_postselected:.16f}, ler = {(1 - ev_postselected):.16f}")

# add borders to fill_between
plt.fill_between(all_p0, -uncertainty_threshold, uncertainty_threshold, color="red", alpha=0.1)
plt.axhline(uncertainty_threshold, xmin=all_p0[0], xmax=all_p0[-1], color="red", alpha=0.5, linewidth=1.5, label=f"Tolerance = {uncertainty_threshold}")
plt.axhline(-uncertainty_threshold, xmin=all_p0[0], xmax=all_p0[-1], color="red", alpha=0.5, linewidth=1.5)

plt.xlabel("Physical error rate ($p$)", fontsize=fontsize)
plt.ylabel("Conditional EV", fontsize=fontsize)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.xscale("log")
plt.margins(x=0)
plt.grid()
plt.legend(fontsize=fontsize_legend + 2)
plt.tight_layout()
plt.savefig(f"{folder}/{"_".join(folder.split("_")[2:])}.pdf")
plt.savefig(f"figures/{"_".join(folder.split("_")[2:])}.pdf")
plt.show()
```
